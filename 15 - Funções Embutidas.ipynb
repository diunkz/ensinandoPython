# Funções embutidas

O interpretador do Python possui várias funções e tipos embutidos que sempre estão disponíveis. A seguir listamos todas as funções em ordem alfabética.



Aqui falaremos de 3 delas: map(), filter() e zip()

## Map

A função map() executa uma função especificada para cada item em um iterável. O item é enviado para a função como um parâmetro.

```python
map(function, algo_iterável)
```

sendo:

**function**, a função a executar em cada item

**algo_iterável**, uma sequencia, coleção, ou objeto iterável. Você pode enviar quantos iteráveis quiser, apenas certifique-se de que a função tenha um parâmetro para cada iterável.

````python
precos = [1000, 1500, 1250, 2500]

def adicionar_imposto(preco):
    return preco * 1.1
````

````python
# utilizando for

precos_com_impostos = []

for preco in precos:
    precos_com_impostos.append(adicionar_imposto(preco))

precos_com_impostos
````

````python
# utilizando map()

precos_com_impostos2 = map(adicionar_imposto, precos)

precos_com_impostos2
````

````python
list(precos_com_impostos2)
````

## Filter

A função filter() retorna um iterador onde os itens são filtrados através de uma função para testar se o item é aceito ou não.

```python
filter(function, algo_iterável)
```

sendo:

**function**, a função a executar em cada item

**algo_iterável**, uma sequencia, coleção, ou objeto iterável. Você pode enviar quantos iteráveis quiser, apenas certifique-se de que a função tenha um parâmetro para cada iterável.

````python
idades = [5, 12, 17, 18, 24, 32]

def verificar(x):
  if x < 18:
    return False
  else:
    return True
````

````python
adultos = filter(verificar, idades)

adultos
````

````python
lista = list(adultos)
lista
````

## Zip

A função zip() retorna um objeto zip, que é um iterador de tuplas onde o primeiro item em cada iterador passado é emparelhado e, em seguida, o segundo item em cada iterador passado é emparelhado, etc.

Se os iteradores passados tiverem comprimentos diferentes, o iterador com menos itens decidirá o comprimento do novo iterador.

```python
zip(iterator1, iterator2, iterator3 ...)
```

````python
a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica", "Vicky")

x = zip(a, b)

x
````

````python
list(x)
````
